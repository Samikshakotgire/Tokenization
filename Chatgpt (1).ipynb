```python
import os
from collections import defaultdict, Counter

def load_text(file_path):
    """Load the text from the given file and convert it to lowercase for case-insensitive analysis."""
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read().lower()  # Case insensitive

def tokenize_text(text):
    """Tokenize the text by splitting it into words."""
    return text.split()

def find_next_words(tokens, initial_set):
    """Find the words that follow the initial set of words in the tokenized text."""
    following_words = []
    initial_set_len = len(initial_set)
    
    for i in range(len(tokens) - initial_set_len):
        # Check if the slice matches the initial set of words
        if tokens[i:i+initial_set_len] == initial_set:
            # Add the word after the initial set to the list of following words
            following_words.append(tokens[i + initial_set_len])
    
    return following_words

def get_top_tokens(following_words, top_n=4):
    """Get the top N most frequent words from the list."""
    if not following_words:
        return None, None  # No words found after the initial set
    
    word_counts = Counter(following_words)
    # Get the top N most common words
    most_common = word_counts.most_common(top_n)
    return [word for word, _ in most_common], word_counts

def predict_next_token(file_path, initial_words):
    """Main function to predict the next token based on initial words."""
    # Load and tokenize the text
    text = load_text(file_path)
    tokens = tokenize_text(text)
    
    # Find the words that follow the initial set of words
    initial_set = initial_words.split()
    following_words = find_next_words(tokens, initial_set)
    
    # If no following words are found, return the most frequent word in the text
    if not following_words:
        print(f"No occurrence of initial set found: {initial_words}. Returning the most frequent word in the entire text.")
        most_common_words, word_counts = get_top_tokens(tokens)
    else:
        # Get the top 4 most common words following the initial set
        most_common_words, word_counts = get_top_tokens(following_words)
    
    # Output the top results and their histogram
    print(f"Top 4 most likely next words: {most_common_words}")
    print(f"Histogram of word frequencies: {word_counts}")

# Set the directory of your text file
file_path = r"C:\Users\DELL\Downloads\Mahabharata_kisari_mohan_ganguly"  # Change this to your text file path

# Example initial words
initial_words = "Arjun was a "  # Set your initial words here

# Run the prediction
predict_next_token(file_path, initial_words)
```

```python
import re
from collections import defaultdict, Counter
import random

# Function to read the text file
def read_text_file(file_path):
    """
    Reads a text file and returns the text content.
    file_path: path to the text file
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Preprocess the text by tokenizing and cleaning
def preprocess_text(text):
    """
    Preprocess the text by converting it to lowercase and tokenizing.
    text: raw text
    Returns: list of words
    """
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and split into words
    words = re.findall(r'\b\w+\b', text)
    return words

# Build the n-gram model
def build_ngram_model(words, n=2):
    """
    Builds an n-gram model as a dictionary of probabilities.
    words: list of words (tokenized text)
    n: size of n-gram (default is bigram)
    """
    ngrams = defaultdict(Counter)
    
    # Create n-grams
    for i in range(len(words) - n):
        ngram = tuple(words[i:i+n])
        next_word = words[i+n]
        ngrams[ngram][next_word] += 1
    
    # Convert counts to probabilities
    for ngram, next_words in ngrams.items():
        total = float(sum(next_words.values()))
        for word in next_words:
            next_words[word] /= total
    
    return ngrams

# Function to predict the next word
def predict_next_word(ngram_model_bigram, ngram_model_trigram, seed):
    """
    Predict the next word based on trigram model with fallback to bigram model.
    ngram_model_bigram: bigram probability model
    ngram_model_trigram: trigram probability model
    seed: initial n-gram (tuple of words)
    Returns: predicted next word
    """
    current_ngram = seed
    
    # Try trigram model first
    if current_ngram in ngram_model_trigram:
        next_word = random.choices(list(ngram_model_trigram[current_ngram].keys()), 
                                   weights=ngram_model_trigram[current_ngram].values())[0]
    elif current_ngram[-1:] in ngram_model_bigram:
        # Fallback to bigram model if trigram fails
        next_word = random.choices(list(ngram_model_bigram[current_ngram[-1:]].keys()), 
                                   weights=ngram_model_bigram[current_ngram[-1:]].values())[0]
    else:
        # No prediction available
        next_word = None
    
    return next_word

# Example usage:
file_path = 'Downloads/Mahabharata_kisari_mohan_ganguly'  # replace with the path to your file

# Step 1: Read and preprocess the text
raw_text = read_text_file(file_path)
words = preprocess_text(raw_text)

# Step 2: Build the bigram and trigram models
ngram_model_bigram = build_ngram_model(words, n=2)
ngram_model_trigram = build_ngram_model(words, n=3)

# Step 3: Predict the next word using a seed phrase
seed = ('krishna', 'dwaipayana')  # Example seed

# Predict the next word
next_word = predict_next_word(ngram_model_bigram, ngram_model_trigram, seed)

print(f"Predicted next word: {next_word}")
```
